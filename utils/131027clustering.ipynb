```python
def load(fname):
    ''' load the file using std open'''
    f = open(fname,'r')

    data = []
    for line in f.readlines():
        #print line
        data.extend(line.replace('\n','').replace('﻿', '').split(', '))
    
    f.close()

    return data

keyword = load('../utils/data/Keyword.txt')
print " ".join(keyword)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
%pylab inline
```

```python
dftemphead = dftemp
dftemphead = StandardScaler().fit_transform(dftemphead)
print 'dftemphead.shape:', dftemphead.shape
```

```python
##############################################################################
# Compute DBSCAN
dbscan = DBSCAN().fit(dftemphead)
core_samples = dbscan.core_sample_indices_
labels = dbscan.labels_
core_samples
```

```python

from sklearn.externals import joblib
#joblib.dump(dbscan, '../data/my_model01.pkl', compress=9)
#model_clone = joblib.load('my_model.pkl')

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

##############################################################################
# Plot result
import pylab as pl


# Black removed and is used for noise instead.
unique_labels = set(labels)

unique_labels
```

```python
colors = pl.cm.Spectral(numpy.linspace(0, 1, len(unique_labels)))

#pl.xlim(xmin=-4, xmax=4)
#pl.ylim(ymin=-4, ymax=4)

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'
        markersize = 6
    class_members = [index[0] for index in numpy.argwhere(labels == k)]
    cluster_core_samples = [index for index in core_samples
                            if labels[index] == k]
    for index in class_members:
        x = dftemphead[index]
        #x?
        if index in core_samples and k != -1:
            markersize = 14
        else:
            markersize = 6
        pl.plot(x[0], x[1], 'o', markerfacecolor=col,
                markeredgecolor='k', markersize=markersize)
```

PCA를 통해 2차원으로 바꾼 후 시도하면 어떨까.

아래는 분산도 그래프를 그리는 함수이다.

```python
from itertools import cycle

def plot_2D(data, target, target_names):
    colors = cycle('rgbcmykw')
    target_ids = range(len(target_names))
    pl.figure()
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1],
                   c=c, label=label)
    pl.legend()
```
