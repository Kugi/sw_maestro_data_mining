### 0-1. user_gender를 사용해서   

 일단 데이터를 불러온다.  

```python
from sklearn.datasets.svmlight_format import load_svmlight_file

X, y = load_svmlight_file('data/minimize/user_gender_1000.dat') # 전부 다 쓰면 너무 크니까...

print X.shape, y.shape
```

우리는 이 데이터의 답을 알고 있으므로, 클러스터링 결과랑 비교해 볼 수 있다.  

```python
from sklearn.cluster import KMeans # 가장 쓰기 쉬운 녀석.

kms = KMeans(n_clusters=2)  # KMeans는 구분할 대상의 분류 갯수를 지정해 주어야 한다.
kms.fit(X)  # 무식하게 전부 넣고 돌려보자.  
```

일단 에러 없이 학습이 된다.  
이로서 svmlinght를 읽은 결과인 (n_sample, n_feature) 모양의 csr_matrix에 대해서 n_feature의 수에 상관 없이 쓸 수 있음을 알 수 있다.  

결과가 어떻게 나왔는지 보자.  

```python
print kms.cluster_centers_
print kms.labels_[:100]
```

결과도 에러 없이 나온다. 문제는 n_feature의 크기가 5만2천을 넘어가는 너무 큰 수라 결과를 시각적으로 표현 할 수가 없다...  
이 문제를 해결하는건 나중으로 미루고, 일단 정답률을 확인해 보자.  

```python
cases = np.zeros((2,2),dtype=int) # 예측남녀 : 실제남녀 조합으로, 4개의 경우의 수가 존재.
for expect, actual in zip(kms.labels_, y):
    cases[expect][actual-1] += 1
    
print cases
```

분류가 되긴 했는데, 성별을 기준으로 된 것 같진 않다...  
(성별을 기준으로 분류되었다면, 대각선 방향으로 값이 크게 나와야한다.)  
나중에 남녀 비율의 차이가 큰 특징만 추려서 넣어보기로 하고, 일단.. 시각적으로 비교를 해보자.  

pylab을 이용해, 시각적으로 나타내려면 특징을 2개로 압축해야 한다.  

```python
from sklearn.decomposition import PCA

Xa = X.toarray()  # PCA는 (n_sample, n_feature) 모양을 지원하지 않는 모양.

pca = PCA(n_components=2, whiten=True)
pca.fit(Xa)
x2 = pca.transform(Xa)

y2 = array([l-1 for l in y.astype(int)])

print x2.shape, y2.shape
```

작동 원리는 모르겠지만, 이러면 2개의 차원으로 압축할 수 있다고 한다. 시간복잡도는 O(n^3)인 듯.

압축한 자료를 학습시켜 보자  

```python
kms2 = KMeans(n_clusters=2)  # KMeans는 구분할 대상의 분류 갯수를 지정해 주어야 한다.
kms2.fit(x2)  # 이번엔 2차원으로 구분된 녀석을 넣어보자.  
```

학습을 시켰으면, 분산도를 그릴 준비가 되되었다. 분산도를 그리는 함수를 짠다.  

```python
from itertools import cycle

def plot_2D(data, target, target_names):
    colors = cycle('rgbcmykw')
    target_ids = range(len(target_names))
    pl.figure()
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1],
                   c=c, label=label)
    pl.legend()
```

결과를 그려본다.  

```python
plot_2D(x2, y2, ["male","female"]) # 원본 데이터
plot_2D(x2, kms2.labels_, ["category0","category1"]) # 클러스터링 한 결과.  
```

원본과 클러스터링 결과를 보면 뭔가 확연히 다름을 알 수 있다. feature로 사용한 어플 설치 여부중에,  
성별을 구별하는데 별 도움이 되지 않는 어플이 있을거라 예측해 볼 수 있다.  
feature Selection이 필요하지 않을까?

The following are two well-known clustering algorithms. Like most unsupervised learning
models in the scikit, they expect the data to be clustered to have the shape `(n_samples, n_features)`:

- `sklearn.cluster.KMeans`: <br/>
    The simplest, yet effective clustering algorithm. Needs to be provided with the
    number of clusters in advance, and assumes that the data is normalized as input
    (but use a PCA model as preprocessor).
- `sklearn.cluster.MeanShift`: <br/>
    Can find better looking clusters than KMeans but is not scalable to high number of samples.
- `sklearn.cluster.DBSCAN`: <br/>
    Can detect irregularly shaped clusters based on density, i.e. sparse regions in
    the input space are likely to become inter-cluster boundaries. Can also detect
    outliers (samples that are not part of a cluster).

Other clustering algorithms do not work with a data array of shape (n_samples, n_features)
but directly with a precomputed affinity matrix of shape (n_samples, n_samples):

- `sklearn.cluster.AffinityPropagation`: <br/>
    Clustering algorithm based on message passing between data points.
- `sklearn.cluster.SpectralClustering`: <br/>
    KMeans applied to a projection of the normalized graph Laplacian: finds
    normalized graph cuts if the affinity matrix is interpreted as an adjacency matrix of a graph.
- `sklearn.cluster.Ward`: <br/>
    Ward implements hierarchical clustering based on the Ward algorithm,
    a variance-minimizing approach. At each step, it minimizes the sum of
    squared differences within all clusters (inertia criterion).
- `sklearn.cluster.DBSCAN`: <br/>
    DBSCAN can work with either an array of samples or an affinity matrix.
