```python
import sklearn.datasets.svmlight_format


# 이보다 아래 있는 _dump_svmlight(X, y, f, one_based, comment, query_id)를 작동시키기 위한 구문.
# github에는 sklearn.externals.six 안에 선언 되어 있다. (현 버전에는 없는 듯 하다.)
# 도대체 배포를 언제 하려는 건지 ㅇ<-<

def _add_doc(func, doc):
    """Add documentation to a function."""
    func.__doc__ = doc

import sys
PY3 = sys.version_info[0] == 3
if PY3: 
    def b(s):
        return s.encode("latin-1")
    def u(s):
        return s
    if sys.version_info[1] <= 1:
        def int2byte(i):
            return bytes((i,))
    else:
        # This is about 2x faster than the implementation above on 3.2+
        int2byte = operator.methodcaller("to_bytes", 1, "big")
    import io
    StringIO = io.StringIO
    BytesIO = io.BytesIO
else:
    def b(s):
        return s
    def u(s):
        return unicode(s, "unicode_escape")
    int2byte = chr
    import StringIO
    StringIO = BytesIO = StringIO.StringIO
_add_doc(b, """Byte literal""")
_add_doc(u, """Text literal""")

def _dump_svmlight(X, y, f, one_based, comment, query_id): # sklearn github에서 가져온 코드.
    is_sp = int(hasattr(X, "tocsr"))
    if X.dtype.kind == 'i':
        value_pattern = u("%d:%d") # 이 부분에 해당하는, int형 자료에 대한 처리가 이전 코드엔 없고.
    else:
        value_pattern = u("%d:%.16g") # 이 부분이 %f로 되어 있어 1.000000와 같이 출력되어 공간이 많이 필요 했다. 

    if y.dtype.kind == 'i':
        line_pattern = u("%d")
    else:
        line_pattern = u("%.16g")

    if query_id is not None:
        line_pattern += u(" qid:%d")
    line_pattern += u(" %s\n")

    if comment:
        f.write(b("# Generated by dump_svmlight_file from scikit-learn %s\n"
                % __version__))
        f.write(b("# Column indices are %s-based\n"
                  % ["zero", "one"][one_based]))

        f.write(b("#\n"))
        f.writelines(b("# %s\n" % line) for line in comment.splitlines())

    for i in range(X.shape[0]):
        s = " ".join([value_pattern % (j + one_based, X[i, j])
                      for j in X[i].nonzero()[is_sp]])
        if query_id is not None:
            feat = (y[i], query_id[i], s)
        else:
            feat = (y[i], s)
        f.write((line_pattern % feat).encode('ascii'))
        
sklearn.datasets.svmlight_format._dump_svmlight = _dump_svmlight # hack        
```

## gender classify  

```python
from sklearn.datasets.svmlight_format import load_svmlight_file

X, y = load_svmlight_file("data/user_gender.dat")
n_samples, n_features = X.shape
print("n_samples=%d" % n_samples)
print("n_features=%d" % n_features)
```

```python
from sklearn.svm import LinearSVC
from sklearn.datasets.svmlight_format import dump_svmlight_file

fitted_linearSVC = LinearSVC(C=0.0070671812739274914).fit(X,y) # C value from LinearSVC.ipynb
# 믿고 그냥 학습!!

unclX, predY = load_svmlight_file("unclassified_user_apps.dat")
predY = fitted_linearSVC.predict(unclX)

dump_svmlight_file(unclX, predY,'profiled_user_gender.dat')

print "done"
```

## location classify

```python
from sklearn.datasets.svmlight_format import load_svmlight_file

provX, provY = load_svmlight_file('data/user_prov.dat')
```

지역정보는 학습해본적이 한번도 없기 때문에 이 자리에서 바로 학습- 검증.  

```python
from sklearn.svm import LinearSVC
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split

n_sample_size = 4000
X_train, X_test, y_train, y_test = train_test_split(provX,provY,test_size=0.25, random_state=0)
X_train_small = X_train[:n_sample_size]
y_train_small = y_train[:n_sample_size]

prams = {
         'C': np.logspace(-3, -2, 100),
}
gcv = GridSearchCV(LinearSVC(),prams,cv=3, n_jobs=-1)

%time _ = gcv.fit(X_train_small, y_train_small)

print gcv.best_params_, gcv.best_score_
```
