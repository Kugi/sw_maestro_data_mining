# SVM 사용하기  

SVM을 사용하기 위해서는 아래와 같은 모듈들이 필요하다. ..... 만 위의 3줄 빼고는 굳이 이 위치에 놓을 필요는 없음..;  

```python
import numpy as np
import pylab as pl
from scipy import sparse

# 사용할 svm
from sklearn.svm import LinearSVC
from sklearn.svm import  SVC
# svm포멧을 읽어오는 것.
from sklearn.datasets import load_svmlight_file
# 10-fold를 직접 구현할 필요 없음. 이미 있음.
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import ShuffleSplit


# Some nice default configuration for plots
pl.rcParams['figure.figsize'] = 10, 7.5
pl.rcParams['axes.grid'] = True
pl.gray()
```

데이터를 불러온다.  

```python
X, y = load_svmlight_file("data/user_gender_10000.dat")
n_samples, n_features = X.shape
print("n_samples=%d" % n_samples)
print("n_features=%d" % n_features)
```

우선은 1만개짜리로 불러왔다. (전체를 사용하면 rbf 커널을 쓸때 너무 느리다.)  

## 최적의 파라미터 찾기!

### 1. ShuffleSplit 사용하기  

적합한 감마 값 찾아보는 코드. ShuffleSplit을 사용한다.  

```python
theC = 46.4158883361 # <-3.5938136638
n_gammas = 10
n_iter = 5
n_size = 500
cv = ShuffleSplit(n_samples, n_iter=n_iter, train_size=n_size, test_size=n_size,random_state=0)
    
train_scores = np.zeros((n_gammas, n_iter))
test_scores = np.zeros((n_gammas, n_iter))
gammas = np.logspace(-5, 0, n_gammas)

for i, gamma in enumerate(gammas):
    for j, (train, test) in enumerate(cv):
        clf = SVC(C=theC, gamma=gamma).fit(X[train], y[train])
        train_scores[i, j] = clf.score(X[train], y[train])
        test_scores[i, j] = clf.score(X[test], y[test])
        
audioPlayer('sound/bgm_ivent7.mp3')
```

결과가 나오는데 대략 10분 정도(n_size=1000 일 때) 소요되기 때문에, 계산이 끝나면 mp3가 재생되도록 했다.  

아래는 위 코드의 결과를 그래프로 그리는 코드 이다.

```python
for i in range(n_iter):
    pl.semilogx(gammas, train_scores[:, i], alpha=0.4, lw=2, c='b')
    pl.semilogx(gammas, test_scores[:, i], alpha=0.4, lw=2, c='g')
pl.ylabel("score for SVC(C=%f, gamma=gamma)"%theC)
pl.xlabel("gamma")
pl.text(gammas[1], 0.5, "Underfitting", fontsize=16, ha='center', va='bottom')
pl.text(gammas[5], 0.5, "Good", fontsize=16, ha='center', va='bottom')
pl.text(gammas[8], 0.5, "Overfitting", fontsize=16, ha='center', va='bottom')
```

지금은 C값을 적절한 것으로 찾아두었기 때문에, 밑의 라벨과 그래프의 모양이 안 맞을수도 있다. (실제로 C = 46.41.. 일때 GOOD자리는 이미 Overfitting 이고)  
아래는 제일 높은 gamma 값을 찾기위해 쓴 코드. 인덱스 잘 조절해가면서 수동으로 찾는다.   

```python
print gammas[2]
print test_scores[2]
```

마찬가지로 C값도 같은 방법으로 최적값을 찾아볼 수 있다.  

```python
gamma = 0.000129154966501 #0.000464158883361 <- 0.000464158883361 <- 0.0016681005372
n_Cs = 10
n_iter = 5
n_size = 500
cv = ShuffleSplit(n_samples, n_iter=n_iter, train_size=n_size, test_size=n_size,
    random_state=0)

train_scores = np.zeros((n_Cs, n_iter))
test_scores = np.zeros((n_Cs, n_iter))
Cs = np.logspace(-5, 5, n_Cs)

for i, C in enumerate(Cs):
    for j, (train, test) in enumerate(cv):
        clf = SVC(C=C, gamma=gamma).fit(X[train], y[train])
        train_scores[i, j] = clf.score(X[train], y[train])
        test_scores[i, j] = clf.score(X[test], y[test])
        
audioPlayer('sound/bgm_ivent7.mp3')        
```

C 값 찾는 것이 gamma 찾는것 보단 빠른 것 같지만, 역시나 느리기 때문에, 음악이 재생되도록 하였다.  

마찬가지로, 아래는 결과 값을 그래프로 나타내는 코드이다.  

```python
for i in range(n_iter):
    pl.semilogx(Cs, train_scores[:, i], alpha=0.4, lw=2, c='b')
    pl.semilogx(Cs, test_scores[:, i], alpha=0.4, lw=2, c='g')
pl.ylabel("score for SVC(C=C, gamma=%f)"%gamma)
pl.xlabel("C")
pl.text(Cs[1], 0.5, "Underfitting", fontsize=16, ha='center', va='bottom')
pl.text(Cs[8], 0.5, "Few Overfitting", fontsize=16, ha='center', va='bottom')
```

음, C의 경우 그래프가 다시 떨어지는걸 보기 힘들어서인진 몰라도, 라벨이랑 그래프 모양이랑 gamma값에 상관 없이 잘 맞는 것 같다.  

아래는 위와 마찬가지로, 수동으로 인덱스를 수정해서 최적의 C값을 찾는 코드이다.  

```python
print Cs
print test_scores[6]
```

위 작업의 성과를 알아보기 위해, 기본 값과 최적화 파라미터로 작동한 결과를 비교해 보았다.  

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

clf = SVC()
clf = clf.fit(X_train,y_train)
a = clf.score(X_train,y_train)
b = clf.score(X_test,y_test) 
print "최적화 전 score-> train:%.8f, test:%.8f"%(a,b)

clf = SVC(C=46.4158883361,gamma=0.000129154966501)
clf = clf.fit(X_train,y_train)
a = clf.score(X_train,y_train)
b = clf.score(X_test,y_test) 
print "최적화 후 score-> train:%.8f, test:%.8f"%(a,b)

# 10000개를 가지고 읽음.
# 최적의 C, gamma 값을 모를때, 기본값으로 하면 a, b = 0.575066666667 0.5644
# C=46.4158883361,gamma=0.000129154966501 일 때, a, b =0.9428 0.8812
```

딱 봐도 좋아진걸 알 수 있다. 결과적으로 88%의 정답률을 가지는 Classifier를 만들었다.  

### 2. GridSearchCV 사용하기  

GridSearchCV라는 것이 있는데, 위의 작업을 자동화 시켜주는 것이다.  
위에서 썼던, 이전의 방법은, C값을 찾고, 그 C값을 gamma값 찾는데 적용해 보고, 나온 gamma값을 다시 C값 찾는데 적용해보고..  
이러한 과정을 여러번 반복해야 매끄러운 결과가 나왔는데, 그걸 알아서 해준다는 것이다.  
(물론 예시로 든 과정 그대로 알고리즘으로 삼진 않겠지만 말이다. )  

```python
from sklearn.grid_search import GridSearchCV
GridSearchCV?
```

```python
from pprint import pprint
svc_params = {
    'C': np.logspace(-5, 3, 10),
    'gamma': np.logspace(-7, -4, 10),
}
pprint(svc_params)
```

GridSearchCV는 param_grid란 파라미터가 있는데, dictionary 혹은 dictionary들의 리스트를 받는다. (위의 코드가 그걸 정해준 셈.)  

```python
n_subsamples = 2000
X_small_train, y_small_train = X_train[:n_subsamples], y_train[:n_subsamples]
```

실행시간이 매우 긴 편이므로, 작은 사이즈로 쪼개서 사용한다. X_train 그대로 쓰면 끝날줄을 모름(....) 아마 이게 사이즈가 7500개 일텐데....  

```python
gs_svc = GridSearchCV(SVC(), svc_params, cv=3, n_jobs=-1)

%time _ = gs_svc.fit(X_small_train, y_small_train)

audioPlayer('sound/bgm_ivent7.mp3')
```

```python
gs_svc.best_params_, gs_svc.best_score_
```

정확도가 83%정도이고, C값이 매우 크게 나오는데, 테스트 사이즈가 2000이여서 그럴수도 있겠다.  

GridSearchCV는 fit한 이후의 가장 좋은 파라미터 상태를 알아서 가지고 있기 때문에, 그대로 최적화된 SVC처럼 써도 된다.  

```python
gs_svc.score(X_test,y_test)
```

대충 위의 83% 비스므리하게 나왔다.  

## 데이터 nomalize 하기.
